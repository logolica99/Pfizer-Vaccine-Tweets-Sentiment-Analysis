## **Loading module and data**

```python
import pandas as pd
import numpy as np
import re 
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from textblob import TextBlob
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix, ConfusionMatrixDisplay
```

```python
text_df=df.drop(['id', 'user_name', 'user_location', 'user_description', 'user_created',
       'user_followers', 'user_friends', 'user_favourites', 'user_verified',
       'date',  'hashtags', 'source', 'retweets', 'favorites',
       'is_retweet'],axis=1)
```

## **Text Preprocessing**

```python
def data_processing(text):
    text = text.lower()
    text = re.sub(r"https\S+|www\S+https\S+", '',text, flags=re.MULTILINE) # removes the urls from the tweet
    text = re.sub(r'\@w+|\#','',text) # removes the hashtags
    text = re.sub(r'[^\w\s]','',text) # removes the punctuations
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
text_df.text = text_df['text'].apply(data_processing)
```

```python
text_df = text_df.drop_duplicates('text')
```

```python
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]

    return data
```

```python
text_df['text'] = text_df['text'].apply(lambda x: stemming(x))
```

```python
text_df.head()
```

```python
#after stemming
print(text_df['text'].iloc[0],'\n')
```

```python
text_df.info()
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_df['polarity'] = text_df['text'].apply(polarity)
```

```python
text_df.head()
```

```python
def sentiment(label):
    if label<0:
        return "Negative"
    elif label==0:
        return "Neutral"
    elif label>0:
        return "Positive"
```

```python
text_df['sentiment'] = text_df['polarity'].apply(sentiment)
```

```python
#bigram language model
vect = CountVectorizer(ngram_range=(1,2)).fit(text_df['text'])
```

```python
feature_names = vect.get_feature_names()
print("Number of features: {}\n".format(len(feature_names)))
print("First 20 features \n {}".format(feature_names[:20]))
```

## **Building a model**

```python
X = text_df['text']
Y = text_df['sentiment']
X=vect.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
print("Size of x_train:", (x_train.shape))
print("Size of y_train:", (y_train.shape))
print("Size of x_test:", (x_test.shape))
print("Size of y_test:", (y_test.shape))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### **Using just logistic regression**

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, logreg_pred))
print("\n")
print(classification_report(y_test, logreg_pred))
```

```python
style.use('classic')
cm = confusion_matrix(y_test, logreg_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=logreg.classes_)
disp.plot()
```

### **Using Grid Search CV**

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.001, 0.01, 0.1, 1, 10]}
grid = GridSearchCV(LogisticRegression(), param_grid)
grid.fit(x_train, y_train)
```

```python
print("Best parameters:", grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
```

```python
logreg_acc = accuracy_score(y_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, y_pred))
print("\n")
print(classification_report(y_test, y_pred))
```
